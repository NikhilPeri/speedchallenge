```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hybrid_training = pd.read_csv('../logs/hybrid.csv')
optical_flow_training = pd.read_csv('../logs/optical_flow.csv')
```

```python
plt.plot(hybrid_training.drop([0,1])['val_mse'], label="Segmentation + Optical Flow MSE")
plt.plot(optical_flow_training.drop([0,1])['val_mse'], label="Optical Flow MSE")
# Add legend
plt.legend(loc='top right')
# Add title and x, y labels
plt.title("Segmentaion + Optical Flow vs Optical Flow", fontsize=16, fontweight='bold')
plt.suptitle("compared on identical unseen validation set", fontsize=10)
plt.xlabel("Epoch")
plt.grid()
plt.ylabel("Validation MSE")
plt.savefig('../figures/model_validation_comparison.png')
plt.show()
```

```python
plt.plot(hybrid_training.drop([0,1])['train_mse'], label="Segmentation + Optical Flow MSE")
plt.plot(optical_flow_training.drop([0,1])['train_mse'], label="Optical Flow MSE")
# Add legend
plt.legend(loc='top right')
# Add title and x, y labels
plt.title("Segmentaion + Optical Flow vs Optical Flow", fontsize=16, fontweight='bold')
plt.suptitle("No significant training variation", fontsize=10)
plt.xlabel("Epoch")
plt.grid()
plt.ylabel("Training MSE")
plt.savefig('../figures/model_training_comparison.png')
plt.show()
```
